2. Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões? [‘mean’, np.median, pd.Series.mod]

```python
import pandas as pd
import numpy as np

# Carregar o arquivo CSV
df = pd.read_csv('SAUDE_SONO.csv')

# Agrupar por profissão e calcular a média, mediana e moda
result = df.groupby('Profissão')['Duração do sono'].agg(
    mean='mean',
    median='median',
    mode=lambda x: pd.Series.mode(x).iloc[0] if not pd.Series.mode(x).empty else np.nan
)

# Arredondar os resultados para duas casas decimais
result = result.round(2)

# Visualizar o resultado
print(result)
```

3. Das pessoas que atuam com engenharia de software qual a porcentagem de obesos?

```python
# Filtrar as pessoas que atuam com engenharia de software
engenharia_software = df[df['Profissão'] == 'Eng. de Software']

# Calcular o número total de engenheiros de software
total_engenheiros = len(engenharia_software)

# Calcular o número de engenheiros de software obesos
obesos_engenheiros = len(engenharia_software[engenharia_software['Categoria IMC'] == 'Obesidade'])

# Calcular a porcentagem de obesos
porcentagem_obesos = (obesos_engenheiros / total_engenheiros) * 100

porcentagem_obesos
```

✅ATIVIDADE Nº 8                                                               
  Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a `média`)                                                                                                                                         
  Portanto, as mulheres têm uma pressão sanguínea média maior do que os homens.

```python
# Primeiro, precisamos converter a coluna 'Pressão Sanguínea' em valores numéricos
# Vamos separar a pressão sistólica e diastólica para calcular a média corretamente

# Função para extrair os valores sistólicos e diastólicos
def extract_pressure(pressure):
    systolic, diastolic = pressure.split('/')
    return int(systolic), int(diastolic)

# Aplicar a função para criar novas colunas
df[['Pressão Sistólica', 'Pressão Diastólica']] = df['Pressão Sanguínea'].apply(lambda x: pd.Series(extract_pressure(x)))

# Calcular a média da pressão sistólica e diastólica por gênero
media_pressao_por_genero = df.groupby('Gênero')[['Pressão Sistólica', 'Pressão Diastólica']].mean()

# Calcular a pressão sanguínea média como a média da sistólica e diastólica
media_pressao_por_genero['Pressão Sanguínea Média'] = media_pressao_por_genero.mean(axis=1)

media_pressao_por_genero
```

Arredondando os valores para dois dígitos decimais


```python
media_pressao_por_genero = media_pressao_por_genero.round(2)

media_pressao_por_genero
```

ATIVIDADE Nº 10                                                           
  Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a `média`)

     Pessoas com frequência cardíaca acima de 70: apresentam uma média diária de 6473 dados.
     Pessoas com frequência cardíaca menor ou igual a 70: apresentam uma média diária de 7007 dados.
     Concluí-se que pessoas com frequência cardíaca menor ou igual a 70 dão mais passos, em média, do que aquelas com frequência cardíaca acima de 70.

```python
# Separar os dados em dois grupos: frequência cardíaca acima de 70 e menor ou igual a 70
frequencia_acima_70 = df[df['Frequência cardíaca'] > 70]
frequencia_menor_igual_70 = df[df['Frequência cardíaca'] <= 70]

# Calcular a média de passos diários para cada grupo
media_passos_acima_70 = frequencia_acima_70['Passos diários'].mean()
media_passos_menor_igual_70 = frequencia_menor_igual_70['Passos diários'].mean()

media_passos_acima_70, media_passos_menor_igual_70
```

```python
# Arredondar as médias de passos para números inteiros
media_passos_acima_70 = round(media_passos_acima_70)
media_passos_menor_igual_70 = round(media_passos_menor_igual_70)
media_passos_acima_70, media_passos_menor_igual_70
```
